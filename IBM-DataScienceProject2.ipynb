```python
def inspect_df(df):
    print('\n','*'*35)
    print('Printing info on DataFrame','\n')
    print('*'*35,'\n')
    df.info()
    
    print('\n','*'*35)
    print('Printing random 5 values','\n')
    print('*'*35,'\n')
    display(df.sample(5,random_state = seed_number))
    
    print('\n','*'*35)
    print('Printing Statistics on DataFrame','\n')
    print('*'*35,'\n')
    display(np.round(df.describe(),2).T)
    
    print('\n','*'*35)
    print('Inspecting missing values on DataFrame','\n')
    print('*'*35,'\n')
    print(df.isna().sum())

    print('\n','*'*35)
    print('Inspecting missing values on DataFrame','\n')
    print('*'*35,'\n')
    msno.matrix(df)
    plt.show()
```

## Modelling

Using LazyRegressor to get a baseline

```python
lazy_model = LazyRegressor(verbose=0, ignore_warnings=True, custom_metric = None)
lazy_models,lazy_models_preds = lazy_model.fit(X_train, X_test, y_train, y_test)
# Viewing only top 10 models
lazy_models[:10]
```

```python
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import xgboost as xgb
```

```python
rf = RandomForestRegressor()

rf_param_grid = {'n_estimators' : (25,50,75) ,
                'max_depth': (10,15,20),
                'min_samples_leaf' :(2,3),
                 'random_state' : [seed_number],
                 'n_jobs' : [-1]
                }

rf_gridsearch = GridSearchCV(estimator = rf, param_grid= rf_param_grid, refit = True , cv = 5)

rf_gridsearch.fit(X_train, y_train)
rf_prediction = rf_gridsearch.predict(X_test)
print('RMSE: \t\t', mean_squared_error(y_test, rf_prediction)**(1/2))
print('MAE: \t\t', mean_absolute_error(y_test, rf_prediction))
print('R2 Score: \t',r2_score(y_test, rf_prediction))
print('Best Params :\t', rf_gridsearch.best_params_)
```

```python
bag_reg = BaggingRegressor()

bag_reg_param_grid = {'n_estimators' : (range(300,450,50)) ,
                      'max_features': (0.5,0.7,0.9,1),
                      'max_samples' :(range(1500,1700,50)), 
                      'random_state' : [seed_number],
                      'n_jobs' : [-1]
                }

bag_reg_gridsearch = GridSearchCV(estimator = bag_reg, param_grid= bag_reg_param_grid, refit = True , cv = 5)

bag_reg_gridsearch.fit(X_train, y_train)
bag_reg_prediction = bag_reg_gridsearch.predict(X_test)
print('RMSE: \t\t', mean_squared_error(y_test, bag_reg_prediction)**(1/2))
print('MAE: \t\t', mean_absolute_error(y_test, rf_prediction))
print('R2 Score: \t',r2_score(y_test, bag_reg_prediction))
print('Best Params :\t', bag_reg_gridsearch.best_params_)
```
